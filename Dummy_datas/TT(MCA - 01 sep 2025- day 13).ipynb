 --- Data cleaning and missing values in pandas ---

Data cleaning is the process of ensuring the quality and the entigrity of the data.

NAN is the technique to find the missing data and the repeated data.

Missing data :- Whenever we have to find missing data ,we use dropna()

Fill the data :- Whenever we have to fill the data , we use fillna()

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = {
    "A":[1,2,np.nan,4,5],
    "B":[np.nan,2,3,4,5],
    "C":[1,2,3,np.nan,5],
    "D":[1,2,3,4,5]
}

df = pd.DataFrame(data)

print(df)
print()
df.fillna(100)
```

 --- Visualization with Pandas --- 
 
--> Ploting data directy from DataFrame using plot() is known as Visualization with pandas 

```python
import matplotlib.pyplot as plt
```

```python
#Sample Data

data = {
    "Month":["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales":[200,250,300,350,300,450],
    "Profit":[50,70,20,52,90,30]
}

#Create DataFrame
df = pd.DataFrame(data)

#Line plot

plt.plot(df["Month"],df["Sales"],label="Sales")
plt.plot(df["Month"],df["Profit"],label="Profit")
plt.xlabel("Month")
plt.ylabel("Values")
plt.title("Sales vs profit (Line Plot)")
plt.legend()
plt.show()


#bar plot
df.plot(x="Month",y=["Sales","Profit"],kind="bar")
plt.title("Sales vs profit (Bar Plot)")
plt.show()
```

```python
#Sample Data

data = {
    "Month":["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales":[200,250,300,350,300,450],
    "Profit":[50,70,20,52,90,30]
}

#Create DataFrame
df = pd.DataFrame(data)

#Line plot

plt.scatter(df["Month"],df["Sales"],label="Sales")
plt.scatter(df["Month"],df["Profit"],label="Profit")
plt.xlabel("Month")
plt.ylabel("Values")
plt.title("Sales vs profit (Scatter Plot)")
plt.legend()
plt.show()
```

```python
#Sample Data

data = {
    "Month":["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales":[200,250,300,350,300,450],
    "Profit":[50,70,20,52,90,30]
}

#Create DataFrame
df = pd.DataFrame(data)

#Line plot

plt.bar(df["Month"],df["Sales"],label="Sales")
plt.bar(df["Month"],df["Profit"],label="Profit")
plt.xlabel("Month")
plt.ylabel("Values")
plt.title("Sales vs profit (Bar Plot)")
plt.legend()
plt.show()
```

```python

#Hisogram

plt.hist(df["Sales"],bins=5,edgecolor = "black")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Sales Distribution (Histogram)")
plt.show()
```

```python

#PieChart

data = {
    "Product" : ["A","B","C","D","E"],
    "Sales" : [3400,4599,5699,1200,3456]
}

df = pd.DataFrame(data)

#Pie Plot
plt.pie(df["Sales"],labels=df["Product"],autopct='%1.1f%%',startangle=90)
plt.legend()
plt.title("Sales Distribution (Pie Plot)")
plt.show()
```

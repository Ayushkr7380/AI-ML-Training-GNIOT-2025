```python
#If you want ot handle some missing values.

df["price"] = pd.to_numeric(df["price"],errors = "coerce")
df["price"] = df["price"].fillna(0)
```

```python
#Reshaping and pivoting Data

#Reshaping  

#Reshaping data uses pivot() and melt() keyword for restructuring datasets.
#It facilitates data aggregation ad analysis in AI-Driven business insights. 


data = {
    "Date":["2025-08-02","2025-08-12","2025-07-20","2025-05-10","2025-05-11"],
    "Category":["A","B","C","D","A"],
    "Value":[10,20,34,12,35]
}

df = pd.DataFrame(data)

print("Original data frame : \n",df)

print()
#Pivot Data Frame

pivot_df = df.pivot(index="Date",columns="Category",values="Value")
print("Reshaped Data Frame : \n",pivot_df)
```
